write video

```python
frame_width = int(cap.get(3)) #3 is for width
frame_height = int(cap.get(4)) #4 is for height

out_mp4 = cv2.VideoWriter('./video/output.mp4', cv2.VideoWriter_fourcc(*'avc1'), 10, (frame_width, frame_height))
```

```python
while(cap.isOpened()):
    ret, frame = cap.read()

    if ret == True:

        out_mp4.write(frame)
    else:
        break
```

```python
cap.release()
out_mp4.release()
```
