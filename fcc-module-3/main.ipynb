```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
from PIL import Image
```

Annotate images

```python
silvia_img = cv2.imread('images/silvia.jpg', cv2.IMREAD_COLOR)
plt.imshow(silvia_img[:, :, ::-1])
```

drawing line

```python
silvia_img2 = silvia_img.copy()
cv2.line(silvia_img2, (200, 200), (400, 200), (255, 0, 0), 5, cv2.LINE_8)
plt.imshow(silvia_img2[:, :, ::-1])
```
