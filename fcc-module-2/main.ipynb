```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

original pixelated image

```python
pix_img = cv2.imread('images/piximg.jpg', 0)

plt.imshow(pix_img, cmap = 'gray')
```

individual pixel

```python
print(pix_img[500,250])
```

Modifying image pixel

```python
pix_img_copy = pix_img.copy()
pix_img_copy[500,300] = 255
pix_img_copy[500,301] = 255
pix_img_copy[500,302] = 255
pix_img_copy[500,303] = 255
pix_img_copy[500,304] = 255
pix_img_copy[500,305] = 255
pix_img_copy[500,306] = 255
pix_img_copy[500,307] = 255

plt.imshow(pix_img_copy, cmap = 'gray')
print(pix_img_copy)
```

cropping image

```python
pix_img_bgr = cv2.imread('images/piximg.jpg', cv2.IMREAD_COLOR)
pix_img_rgb = pix_img_bgr[:, :, ::-1]
plt.imshow(pix_img_rgb)
```

```python
cropped_piximg = pix_img_rgb[100:750, 100:1000]
plt.imshow(cropped_piximg)
```

Resize image

Method 1: using Fy and Fx

```python
cloud_img_2x = cv2.resize(cropped_piximg, None, fx = 2, fy = 2)
plt.imshow(cloud_img_2x)
```

Method 2: Specifying exact size

```python
desired_width = 300
desired_height = 300
dim = (desired_width, desired_height)

cloud_img_300x = cv2.resize(cropped_piximg, dim)
plt.imshow(cloud_img_300x)
```

Method 3:Resize while maintain aspect ratio

```python
desired_width = 300
aspect_ratio = desired_width / cropped_piximg.shape[1]
desired_height = int(cropped_piximg.shape[0] * aspect_ratio)
dim = (desired_width, desired_height)

cloud_img_300x = cv2.resize(cropped_piximg, dim)
plt.imshow(cloud_img_300x)
```
